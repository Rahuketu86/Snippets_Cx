###Applying python style enumerate (iterator example)


```python
%%cscexy read_by_lines_enumerate.cs "C:\rahuketu\Courses\scicomp\1 - 1 - Couse Preview (0_57).srt"
using System;
using System.Collections.Generic;

namespace Py
{
    class Index_Tuple<T>
    {
        public int Index{get; private set; }
        public T Value{get; private set; }
        
        public Index_Tuple(int index, T value)
        {
            Index = index;
            Value = value;
        }
    }
    
    static class Utils
    {
        public static IEnumerable<Index_Tuple<T>> enumerate<T>(IEnumerable<T> objs)
        {
            int index = 0;
            foreach(T obj in objs)
            {
                yield return new Index_Tuple<T>(index++, obj);
            }
        }
        
       // public static IEnumerable<T> enumerate2<T>(IEnumerable<T> objs)
       // {
       //     return objs.Select((x,i) => new { Value = x, Index=i });
       // }
    }
    //foreach ( var it in someCollection.Select((x,i) => new { Value = x, Index=i }) )
    //{
    //   if ( it.Index > SomeNumber) //      
    //}
}

namespace Playground
{
    class Read_All_By_Lines
    {
        public static void Main(string[] args)
        {
            string[] lines = System.IO.File.ReadAllLines(@args[0]);
            Console.WriteLine("contents.length = {0}", lines.Length);

            foreach(var tuple in Py.Utils.enumerate(lines))
            {
                Console.WriteLine(tuple.Index + "::" + tuple.Value);
            }
            
        }
    }
}
```

###Declarative Linqish way  | So Nice :)  |

```python
%%cscexy read_by_lines_linqish.cs "C:\rahuketu\Courses\scicomp\1 - 1 - Couse Preview (0_57).srt"
using System;
using System.Collections.Generic;

namespace Linqish
{   
    static class Utils
    {
        public static void Each<T>(this IEnumerable<T> objs, Action<T, int> action)
        {
            var i = 0;
            foreach(var e in objs) action(e, i++);
        }
    }
}

namespace Playground
{
    using Linqish;
    class Read_All_By_Lines
    {
        public static void Main(string[] args)
        {
            string[] lines = System.IO.File.ReadAllLines(@args[0]);
            Console.WriteLine("contents.length = {0}", lines.Length);
            lines.Each( (line, n) =>
                        {
                            Console.WriteLine(n + "::" + line);
                        });
            Console.WriteLine("List of Arguments");
            args.Each( (line, n) =>
                       {
                            Console.WriteLine(n + "::" + line);
                       });
        }
    }
}
```
